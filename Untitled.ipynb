# US Mass Shootings Analysis

```python
import folium
import seaborn as sb
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# Reading the data set
df = pd.read_csv('diff analysis/neww.csv')
```

```python
#A bit of data cleaning
df['places'] = df['places'].replace('workplace', 'Workplace')
df.replace('Male', 'M', inplace = True)
df.replace('Female', 'F', inplace = True)
df.replace('Male & Female', 'M & F', inplace = True)
df.replace('white', 'White', inplace = True)
df.replace('black', 'Black', inplace = True)
df.replace('-', np.NaN, inplace = True)
df.head()
```

# Visualize mass shootings on the U.S map

```python
import pandas as pd
import folium
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

# Add geolocation data for each location
geolocator = Nominatim(user_agent="my-custom-user-agent")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
df['location_data'] = df['location'].apply(geocode)

# Extract latitude and longitude data from the geolocation data
df['latitude'] = df['location_data'].apply(lambda location: location.latitude if location else None)
df['longitude'] = df['location_data'].apply(lambda location: location.longitude if location else None)

# Create a map centered on the first location
location = df.iloc[0]['location_data']
map = folium.Map(location=[location.latitude, location.longitude], zoom_start=10)

# Add markers for each location
for index, row in df.iterrows():
    location = row['location_data']
    if location:
        folium.Marker(location=[location.latitude, location.longitude]).add_to(map)

# Display the map
map
```

# Is there any correlation between shooter and his/her race, gender


```python
sb.countplot(x=df['gender'])
```

```python
#Grouping the age into Minor, Teenager, Young Adult, Adult, and Elder
def age_group(x):
    if x <= 12:
        return 'Minor'
    elif x <= 19:
        return 'Teenager'
    elif x <= 30:
        return 'Young adult'
    elif x <=50:
        return 'Adult'
    else:
        return 'Elder'
```

```python
df['US shooting Age_grp'] = df['age_of_shooter'].apply(age_group)
df.head()
```

```python
sb.countplot(x=df['US shooting Age_grp'])
```

```python
# Create a cross-tabulation table of shooter's race, gender, and age group
crosstab = pd.crosstab([df['race'], df['gender']], df['US shooting Age_grp'])
```

```python
# Create a heatmap of the cross-tabulation table
sb.heatmap(crosstab, cmap='coolwarm', annot=True, fmt='d')
# Display the heatmap
plt.show()
```

### The resulting heatmap will show the frequency of shooters based on their race, gender, and age group. Darker colors indicate higher frequency, while lighter colors indicate lower frequency.

# What cities and states are more prone to such attacks


```python
# separating the location into city and state
df['city'] = ""
df['state'] = ""

# Split location into city and state
for index, row in df.iterrows():
    if not pd.isna(row['location']):
        loc_parts = row['location'].split(',')
        if len(loc_parts) == 2:
            df.at[index, 'city'] = loc_parts[0].strip()
            df.at[index, 'state'] = loc_parts[1].strip()
```

```python
df.head()
```

```python
# Group the shootings by city and state
shootings_by_city = df.groupby('city')['total_victims'].count().sort_values(ascending=False)
shootings_by_state = df.groupby('state')['total_victims'].count().sort_values(ascending=False)

# Print the top 10 cities and states with the most shootings
print("Top 10 Cities with the Most Shootings:")
print(shootings_by_city.head(10))
print("\nTop 10 States with the Most Shootings:")
print(shootings_by_state.head(10))
```

```python

# Group the shootings by city and state
shootings_by_city = df.groupby('city')['date'].count().sort_values(ascending=False)
shootings_by_state = df.groupby('state')['date'].count().sort_values(ascending=False)

# Print the top 10 cities and states with the most shootings
print("Top 10 Cities with the Most Shootings:")
print(shootings_by_city.head(10))
print("\nTop 10 States with the Most Shootings:")
print(shootings_by_state.head(10))
```

```python

# Group the shootings by city and state
shootings_by_city = df.groupby('city')['weapon_type'].count().sort_values(ascending=False)
shootings_by_state = df.groupby('state')['weapon_type'].count().sort_values(ascending=False)

# Print the top 10 cities and states with the most shootings
print("Top 10 Cities with the Most Shootings:")
print(shootings_by_city.head(10))
print("\nTop 10 States with the Most Shootings:")
print(shootings_by_state.head(10))
```

# Correlation between Fatalities/Injuries and Time:

```python
# convert 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# extract day, month, and year components
df['year'] = df['date'].dt.year
df['day'] = df['date'].dt.day
df['month'] = df['date'].dt.month
df.head(1)
```

```python
# Create new columns for year and month
df['year'] = pd.to_datetime(df['date']).dt.year
df['month'] = pd.to_datetime(df['date']).dt.month

# Calculate correlation coefficients
corr_year = df['year'].corr(df['fatalities'])
corr_month = df['month'].corr(df['fatalities'])

print(corr_year)
print(corr_month)

```

```python
# convert the date column to datetime format
df['Date'] = pd.to_datetime(df['date'], format='%m/%d/%Y')

# group the data by year and calculate the total fatalities and injuries
yearly_totals = df.groupby(df['date'].dt.year)[['fatalities', 'injured']].sum()

# plot the yearly totals
yearly_totals.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Total Fatalities and Injuries')
plt.title('Mass Shooting Fatalities and Injuries by Year')
plt.show()
```

# Correlation between shooter race/gender and number of victim

```python
# group the data by shooter race and calculate the average number of fatalities and injuries
race_totals = df.groupby('race')[['fatalities', 'injured']].mean()

# plot the average number of fatalities and injuries by shooter race
race_totals.plot(kind='bar')
plt.xlabel('Shooter Race')
plt.ylabel('Average Fatalities and Injuries')
plt.title('Mass Shooting Fatalities and Injuries by Shooter Race')
plt.show()

# group the data by shooter gender and calculate the average number of fatalities and injuries
gender_totals = df.groupby('gender')[['fatalities', 'injured']].mean()

# plot the average number of fatalities and injuries by shooter gender
gender_totals.plot(kind='bar')
plt.xlabel('Shooter Gender')
plt.ylabel('Average Fatalities and Injuries')
plt.title('Mass Shooting Fatalities and Injuries by Shooter Gender')
plt.show()
```

# Correlation between mental health issues and number of victims:

```python
# calculate the number of shooters with mental health issues and the number of fatalities and injuries they caused
mental_health_totals = df.groupby('prior_signs_mental_health_issues')[['fatalities', 'injured']].sum()

# plot the total number of fatalities and injuries caused by shooters with mental health issues
mental_health_totals.plot(kind='bar')
plt.xlabel('Shooter has Mental Health Issues')
plt.ylabel('Total Fatalities and Injuries')
plt.title('Mass Shooting Fatalities and Injuries by Mental Health Issues')
plt.show()
```

# Correlation between location and number of victims:

```python
# calculate the average number of fatalities and injuries for shootings that occurred indoors vs. outdoors
location_totals = df.groupby('places')[['fatalities', 'injured']].mean()

# plot the average number of fatalities and injuries by location
location_totals.plot(kind='bar')
plt.xlabel('Shooting Location')
plt.ylabel('Average Fatalities and Injuries')
plt.title('Mass Shooting Fatalities and Injuries by Location')
plt.show()
```
